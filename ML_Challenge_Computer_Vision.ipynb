```python
## train, test 데이터 csv 파일 업로드 ##
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
import pandas as pd
train_data = pd.read_csv('C:/Users/jkseo/Desktop/challenge/train_vision.csv')
test_data = pd.read_csv('C:/Users/jkseo/Desktop/challenge/test_vision.csv')
```

```python
train_data
```

```python
tr = train_data.values
print(tr)
```

```python
## 파일 이름과 라벨을 각각 추출 ##
import keras

train_file_name = tr[0:, 0]
train_labels = tr[0:, 1] - 1
print(train_file_name)
print(train_labels)
print(train_file_name.shape)
print(train_labels.shape)
## 원핫 인코딩 ##
one_hot_labels = keras.utils.to_categorical(train_labels)
print(one_hot_labels)
print(one_hot_labels.shape)
y_train = one_hot_labels
```

```python
## train_vision 파일에 존재하는 이미지 경로 재설정 ##
import shutil

src = 'C:/Users/jkseo/Desktop/challenge/faces_images/'
dir = 'C:/Users/jkseo/Desktop/challenge/train_data/'

for i in range(len(train_file_name)):
    filename = train_file_name[i]
    shutil.move(src + filename, dir + filename)
```

```python
## train image로 구성된 train dataset 구성 ##
import numpy as np
from keras.preprocessing import image
train_image = []

path = 'C:/Users/jkseo/Desktop/challenge/train_data'
for i in range(len(train_file_name)):
    img = image.load_img(path + '/' + train_file_name[i], target_size=(128, 128))
    img = image.img_to_array(img)
    img = img/255
    train_image.append(img)
    
x_train = np.array(train_image)
```

```python
## 이미지 학습을 위한 CNN 모델 구축 ##
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), input_shape=(128, 128, 3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(64, (3, 3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(128, (3, 3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Conv2D(256, (3, 3)))
model.add(layers.Activation('relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2)))

model.add(layers.Flatten()) 
model.add(layers.Dense(512))
model.add(layers.Activation('relu'))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(6, activation="softmax"))
```

```python
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer='adam', metrics=['acc'])
```

```python
model.fit(x_train, y_train, batch_size=32, epochs=20)
```

```python
test = test_data.values
print(test)

test_file_name = test[0:, 0]
print(test_file_name)
```

```python
## 정확도 검증을 위한 test dataset 구성 ##
path = 'C:/Users/jkseo/Desktop/challenge/faces_images/faces_images'
test_image = []
for i in range(len(test_file_name)):
    img = image.load_img(path + '/' + test_file_name[i], target_size=(128, 128))
    img = image.img_to_array(img)
    img = img/255
    test_image.append(img)
test_img = np.array(test_image)
print(len(test_img))
```

```python
## 예측 결과 확인 ##
prediction = model.predict_classes(test_img)
print(prediction)
```
